```python
# Importing Modules
import os
import wikipedia
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Initialize Chrome Options
chrome_options = Options()
chrome_path = r"C:\Users\aditi\OneDrive\Desktop\chromedriver.exe"
chrome_options.add_argument(f"webdriver.chrome.driver={chrome_path}")
driver = webdriver.Chrome(options=chrome_options)
```

```python
#Lists
monument=[] #list of monuments
hyphen_list=[] #list with hyphen between words and lowercase 
empty_info=[] #list of monuments with empty info

#get the the list of monuments through subdirectories
def get_subdirectories(directory):
    subdirectories = [d for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))]
    return subdirectories
directory_path = "monuments/mumbai/testing"
subdirectories_list = get_subdirectories(directory_path)

for i in subdirectories_list[1:]:
    monument.append(i)    

hyphen_list = [word.replace(' ', '-').lower() for word in monument]
print(hyphen_list)
```

```python
#WIKI Search
def wiki_monument_info(monument_name):
    try:
        page = wikipedia.page(monument_name)
        title = page.title
        summary = page.summary
        url = page.url
        print(title + "\n" + url + "\n" + summary)
        return {"title": title, "summary": summary, "url": url}        
    except wikipedia.exceptions.PageError:
        return {"error": f"Page for '{monument_name}' not found."}
    except wikipedia.exceptions.DisambiguationError as e:
        return {"error": f"Multiple pages found for '{monument_name}'. Suggestions: {e.options}"}
```

```python
def get_monument_info(word,class_):
    url=f"https://www.thrillophilia.com/attractions/{word}"
    driver.get(url)
    content = driver.page_source
    soup = BeautifulSoup(content, "html.parser")
    info = ""
    try:
        # Find all div tags with class 'RawHTML_rawHtmlContainer__99WNn'
        adivs = soup.find('div', class_=class_)
        divs = soup.find_all('div', class_=class_)
        # Extract text from the first four paragraphs inside each div
        for div in divs:
            paragraphs = div.find_all('p')[:4]
            for paragraph in paragraphs:
                info += paragraph.get_text() + "\n\n"
            if len(paragraphs)==0:
                for i in adivs.contents:
                    if i.name == 'h2' or i.name=='b' or i.name == 'h3':
                        break
                    info+= str(i)

    except Exception as e:
        print(f"Error: {e}")

    # Check if info is empty and search info through wiki
    if not info.strip():
        monument_name = word.replace('-',' ')
        monument_info = wiki_monument_info(monument_name)
        if "error" in monument_info:
            print(monument_info["error"])
    else:
        # Print the extracted information
        print(word.upper())
        print(url)
        print(info.replace("<br/>","\n"))   
```

```python
class_='RawHTML_rawHtmlContainer__99WNn DescriptionBox_descriptionRawHTML__16MWf'
for word in hyphen_list:
    get_monument_info(word,class_)
# Close the webdriver
driver.quit()
```

```python
class_='RawHTML_rawHtmlContainer__99WNn DescriptionBox_descriptionRawHTML__16MWf'
cities=["mumbai","jodhpur","delhi","jaipur","udaipur","aurangabad"]
for i in cities:
    dir_path=f"monuments/{i}/training"
    subdir_list=get_subdirectories(dir_path)
    monuments=[]
    hyphen_list=[]
    for j in subdir_list[1:]:
        monuments.append(j)
    hyphen_list = [word.replace(' ', '-').lower() for word in monuments]
    for word in hyphen_list:
        get_monument_info(word,class_)
# Close the webdriver
driver.quit()
```
